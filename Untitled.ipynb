```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV files
at_df = pd.read_csv('at.csv')
iperf_df = pd.read_csv('totaliperf.csv')
ping_df = pd.read_csv('ping.csv')
```

```python
# Remove any rows with extreme values for each file
ping_df = ping_df[ping_df['Latency'] < 100]

# ipfer_df does not have a specific criteria mentioned, so we will skip it for now.

# Rsrp file
at_df = at_df[(at_df['PRX path RSRP value (dBm)'] > -140) & (at_df['PRX path RSRP value (dBm)'] < -44) &
              (at_df['DRX path RSRP value (dBm)'] > -140) & (at_df['DRX path RSRP value (dBm)'] < -44) &
              (at_df['RX2 path RSRP value (dBm)'] > -140) & (at_df['RX2 path RSRP value (dBm)'] < -44) &
              (at_df['RX3 path RSRP value (dBm)'] > -140) & (at_df['RX3 path RSRP value (dBm)'] < -44) &
              (at_df['PRX path RSRQ value (dB)'] > -20) & (at_df['PRX path RSRQ value (dB)'] < -3) &
              (at_df['DRC path RSRQ value (dB)'] > -20) & (at_df['DRC path RSRQ value (dB)'] < -3) &
              (at_df['RX2 path RSRQ value (dB)'] > -20) & (at_df['RX2 path RSRQ value (dB)'] < -3) &
              (at_df['RX3 path RSRQ value (dB)'] > -20) & (at_df['RX3 path RSRQ value (dB)'] < -3) &
              (at_df['PRX path SINR value (dB)'] > -23) & (at_df['PRX path SINR value (dB)'] < 40) &
              (at_df['DRX path SINR value (dB)'] > -23) & (at_df['DRX path SINR value (dB)'] < 40) &
              (at_df['RX2 path SINR value (dB)'] > -23) & (at_df['RX2 path SINR value (dB)'] < 40) &
              (at_df['RX3 path SINR value (dB)'] > -23) & (at_df['RX3 path SINR value (dB)'] < 40)]

# Calculate the average values for each file
ping_avg = ping_df.groupby('Measurement').mean()
iperf_avg = iperf_df.groupby('measurement').mean()
at_avg = at_df.groupby('Measurement').mean()

# Convert bitrate from bits to Mbits
iperf_avg['bitrate'] = iperf_avg['bitrate'] / 1000000

# Combine the dataframes into one
df = pd.concat([ping_avg, iperf_avg, at_avg], axis=1)
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the correlation matrix using seaborn's heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot the average ping times for each measurement
plt.plot(ping_avg.index, ping_avg['Latency'])
plt.title('Average Ping Times')
plt.xlabel('Measurement')
plt.ylabel('Time (ms)')
plt.show()
```

```python
# Plot the average RSRP for each measurement
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(at_avg.index, at_avg['PRX path RSRP value (dBm)'])
axs[0, 0].set_title('Average PRX path RSRP')
axs[0, 0].set_xlabel('Measurement')
axs[0, 0].set_ylabel('dBm')

axs[0, 1].plot(at_avg.index, at_avg['DRX path RSRP value (dBm)'])
axs[0, 1].set_title('Average DRX path RSRP')
axs[0, 1].set_xlabel('Measurement')
axs[0, 1].set_ylabel('dBm')

axs[1, 0].plot(at_avg.index, at_avg['RX2 path RSRP value (dBm)'])
axs[1, 0].set_title('Average RX2 path RSRP')
axs[1, 0].set_xlabel('Measurement')
axs[1, 0].set_ylabel('dBm')

axs[1, 1].plot(at_avg.index, at_avg['RX3 path RSRP value (dBm)'])
axs[1, 1].set_title('Average RX3 path RSRP')
axs[1, 1].set_xlabel('Measurement')
axs[1, 1].set_ylabel('dBm')

plt.tight_layout()
plt.show()
```

```python
# Plot the average RSRQ for each measurement
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(at_avg.index, at_avg['PRX path RSRQ value (dB)'])
axs[0, 0].set_title('Average PRX path RSRQ')
axs[0, 0].set_xlabel('Measurement')
axs[0, 0].set_ylabel('dB')

axs[0, 1].plot(at_avg.index, at_avg['DRC path RSRQ value (dB)'])
axs[0, 1].set_title('Average DRC path RSRQ')
axs[0, 1].set_xlabel('Measurement')
axs[0, 1].set_ylabel('dB')

axs[1, 0].plot(at_avg.index, at_avg['RX2 path RSRQ value (dB)'])
axs[1, 0].set_title('Average RX2 path RSRQ')
axs[1, 0].set_xlabel('Measurement')
axs[1, 0].set_ylabel('dB')

axs[1, 1].plot(at_avg.index, at_avg['RX3 path RSRQ value (dB)'])
axs[1, 1].set_title('Average RX3 path RSRQ')
axs[1, 1].set_xlabel('Measurement')
axs[1, 1].set_ylabel('dB')

plt.tight_layout()
plt.show()
```
